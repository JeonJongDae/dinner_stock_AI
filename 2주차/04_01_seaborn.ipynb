# 4.1.8 히트맵

```python
pivot_df = df.pivot_table("tip", "day", "size")
pivot_df
```

-----
<div class="alert alert-warning">
[Tip] 피봇 테이블의 값 연산 변경
</div>

```python
# 피봇테이블의 값을 개수로 지정
df.pivot_table("tip", "day", "size", aggfunc='count')
```

----------

```python
sns.heatmap(pivot_df)
```

```python
sns.heatmap(pivot_df, annot=True)
```
